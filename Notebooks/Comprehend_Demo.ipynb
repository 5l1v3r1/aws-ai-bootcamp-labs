```python
def getSentiment(inputText, languageCode ):
    textSentiment = comprehend.detect_sentiment(
        Text=inputText,
        LanguageCode=languageCode
    )
    print (inputText)
    print ('Sentiment: %s' % textSentiment['Sentiment'])
    for all in textSentiment['SentimentScore']:
        print (all, textSentiment['SentimentScore'][all]) 
    print (' ')  
```
